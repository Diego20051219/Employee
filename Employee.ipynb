```python
 from abc import ABC, abstractmethod

# Abstract Employee
class AbstractEmployee(ABC):
    def __init__(self):
        self.id = AbstractEmployee.new_id
        AbstractEmployee.new_id += 1
    @abstractmethod
    def say_id(self):
        pass

# Employee
class Employee(AbstractEmployee):
    new_id = 1

    def __init__(self):
        self.id = Employee.new_id
        Employee.new_id += 1
        self._name = None  # Atributo para el nombre del empleado
        self.__company = "Compañia Copeer"  # Atributo privado con nombre de la compañía

    def say_id(self):
        print(f"My id is {self.id}")

    # Getter y setter para el atributo _name
    def get_name(self):
        return self._name

    def set_name(self, name):
        self._name = name

    # Deleter para el atributo _name
    def del_name(self):
        del self._name

# User
class User:
    def __init__(self, username, role):
        self.username = "juan"
        self.role = role

    def say_user_info(self):
        print(f"Username: {self.username}, Role: {self.role}")

# Admin
class Admin(Employee, User):
    def __init__(self, username):
        # Inicializa Employee y User
        Employee.__init__(self)
        User.__init__(self, username, "Admin")

    def say_id(self):
        super().say_id()  # Llama a say_id de Employee
        print("I am an Admin")

# Manager
class Manager(Admin):
    def __init__(self, username):
        super().__init__(username)

    def say_id(self):
        super().say_id()  # Llama a say_id de Admin
        print("I am in charge")

# Meeting
class Meeting:
    def __init__(self):
        self.attendees = []

    def __add__(self, employee):
        if isinstance(employee, Employee):
            self.attendees.append(employee)
            print(f"Empleado con ID {employee.id} agregado exitosamente.")
        return self

    def __len__(self):
        return len(self.attendees)

# Crear instancias de Employee, Admin y Manager
e1 = Employee()
e2 = Employee()
e3 = Admin("admin_username")
e4 = Manager("manager_username")

# Llamar say_id en e1, e2, e3 y e4
e1.say_id()
e2.say_id()
e3.say_id()
e4.say_id()

# Crear una reunión con los empleados
meeting = [e1, e2, e3, e4 ]

# Hacer que todos digan su ID
for attendee in meeting:
    attendee.say_id()

# Crear instancia de Meeting y agregar empleados
m1 = Meeting()
m1 += e1
m1 += e2
m1 += e3

# Imprimir el tamaño de la instancia de Meeting
print(f"El tamaño de la instancia m1 es {len(m1)}")

# Usar dir() para mostrar todos los atributos de una instancia de Employee
print(dir())
```
